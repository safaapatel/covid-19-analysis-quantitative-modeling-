```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame()
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/WidhyaOrg/datasets/master/covid19.csv")
```

```python
df.drop(['Sno','State/UnionTerritory'],axis=1, inplace=True)
pp = df.groupby('Date',sort=False).count()
df1 = pd.DataFrame(pp)
```

```python
df2= df.groupby("Date",sort=False).sum()
df2 = pd.DataFrame(df2)
```

```python
df2
```

```python
df2['Total'] = df2.sum(axis=1)
```

```python
df2
```

```python
df2.drop(['ConfirmedIndianNational','ConfirmedForeignNational','Cured','Deaths'],axis=1).plot.line(title='Covid19 Dataset')
plt.xlabel("Dates")
plt.ylabel("Total cases")
```
